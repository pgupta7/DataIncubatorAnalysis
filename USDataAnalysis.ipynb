```python
import pandas as pd
import numpy as np
import glob
import datetime
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('EdStatsData.csv')
```

```python
df11=df.loc[df['Country Name']=='United States']
df12=df11.loc[df11['Indicator Code']=='SL.UEM.TOTL.FE.ZS']
titleS=df12['Indicator Name'].values[0]
# df12.loc[:,'1970':'2005']
```

```python
df13=pd.DataFrame(df12.iloc[0].values,df12.iloc[0].index)
df13=df13[4:-1]
df13[0]=pd.to_numeric(df13[0],errors='coerce')
df14=df13.loc[~pd.isna(df13[0])]
df14.count()
# pd.to_numeric(df14.columns,errors='coerce')
```

```python
plt.figure()
a=df14.plot(lw=2,figsize=(20,7),title=titleS,fontsize=14)
plt.xticks(list(range(26)), df14.index, fontsize=12)
plt.ylabel('Unemployment percentage',fontsize=14)
plt.xlabel('year',fontsize=14)
plt.show()
```

```python
df_1=pd.read_csv('EdStatsFootNote.csv')
df_1=df_1.rename(index=str, columns={"CountryCode": "Country Code", "SeriesCode": "Indicator Code"})
df_1.head()
```

```python
df_2=df.append(df_1)
```

```python
df_2.head
```

```python
df11=df_2.loc[df_2['Country Code']=='USA']
df11=df11.loc[df11['Indicator Code']=='SE.TER.GRAD.AG.FE.ZS']
df11=df11.loc[~(df11['Indicator Name'].isnull())]
df12=pd.DataFrame(df11.iloc[0].values,df11.iloc[0].index)
df12=df12[:-8]
```

```python
plt.figure()
a=df12.plot(lw=2,figsize=(20,7),title=titleS,fontsize=14)
plt.xticks(list(range(df12.size)), df12.index, fontsize=14)
plt.ylabel('Percentage of youths over 15',fontsize=14)
plt.xlabel('year',fontsize=14)
plt.xticks( rotation='vertical')
plt.show()
```

```python
indicatorCode=['SE.TER.GRAD.ED.FE.ZS',
'SE.TER.GRAD.HU.FE.ZS',
'SE.TER.GRAD.SC.FE.ZS',
'SE.TER.GRAD.SS.FE.ZS']
plt.figure(figsize=(20,10))
comb=pd.DataFrame([])
for i in range(len(indicatorCode)):
    df11=df_2.loc[df_2['Country Code']=='USA']
    df11=df11.loc[df11['Indicator Code']==indicatorCode[i]]
    df11=df11.loc[~(df11['Indicator Name'].isnull())]
    df12=pd.DataFrame(df11.iloc[0].values,df11.iloc[0].index)
    descrip=df11['Indicator Name'].values[0]
    df12=df12[:-8]
    df12=df12.rename(index=str, columns={0:descrip})
    comb=pd.concat([comb,df12],axis=1,sort=False)
    ax=plt.plot(df12, label=descrip)
#     plt.legend(df12['Description'].values, )

plt.xticks(list(range(30,46,1)), df12.index[30:46], fontsize=14)
plt.ylabel('Percentage',fontsize=20)
plt.xlabel('year',fontsize=20)
plt.xticks( rotation='vertical')
plt.yticks( fontsize=14)
plt.legend(fontsize=14,loc='best',bbox_to_anchor=(0.5, 0.5, 0.5, 0.75))
plt.show()
```
