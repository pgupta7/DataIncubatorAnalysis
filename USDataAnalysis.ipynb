```python
import pandas as pd
import numpy as np
import glob
import datetime
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('EdStatsData.csv')
```

```python
df.describe()
```

```python
df11=df.loc[df['Country Name']=='United States']
df12=df11.loc[df11['Indicator Code']=='SL.UEM.TOTL.FE.ZS']
titleS=df12['Indicator Name'].values[0]
# df12.loc[:,'1970':'2005']
```

```python
df13=pd.DataFrame(df12.iloc[0].values,df12.iloc[0].index)
df13=df13[4:-1]
df13[0]=pd.to_numeric(df13[0],errors='coerce')
df14=df13.loc[~pd.isna(df13[0])]
df14.count()
# pd.to_numeric(df14.columns,errors='coerce')
```

```python
plt.figure()
a=df14.plot(lw=2,figsize=(20,7),title=titleS,fontsize=14)
plt.xticks(list(range(26)), df14.index, fontsize=12)
plt.ylabel('Unemployment percentage',fontsize=14)
plt.xlabel('year',fontsize=14)
plt.show()
```
